```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("Unemployment in India.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df.columns =['State','Date','Frequency','Estimated Unemployment Rate','Estimated Employed','Estimated Labour Participation Rate','Region','Area']
```

```python
df['Date']=pd.to_datetime(df['Date'], dayfirst=True)
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.describe()
```

```python
plt.figure(figsize=(10,5))
sns.lineplot(x='Date',y='Estimated Unemployment Rate', data=df)
plt.title("Unemployment Rate in India Over Time")
plt.xlabel("Date")
plt.ylabel("Unemployment Rate(%)")
plt.show()
```

```python
state_unemp=df.groupby('State')['Estimated Unemployment Rate'].mean().sort_values(ascending=False)
```

```python
plt.figure(figsize=(10,6)) 
state_unemp.head(10).plot(kind='bar')
plt.title("Top 10 States by Average Unemployment Rate")
plt.ylabel("unemployment Rate(%)")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='Region',y='Estimated Unemployment Rate',data=df)
plt.title("Region-wise Unemployment Rate")
plt.show()
```

```python
df['Covid_Period']=df['Date'].apply(lambda x:'Covid' if x.year >= 2020  else 'Pre-Covid')
```

```python
sns.barplot(x='Covid_Period',y='Estimated Unemployment Rate', data=df)
plt.title("Unemployment Rate: Pre-Covid Vs Covid")
plt.show()
```

```python
sns.scatterplot(x='Estimated Labour Participation Rate', y='Estimated Unemployment Rate',data=df)
plt.title("Labour Participation vs Unemployment")
plt.show()
```
